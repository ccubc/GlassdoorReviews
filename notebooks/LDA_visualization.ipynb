## Visualize topics for reviews on pros

```python
import gensim
import pickle
import pyLDAvis
import pyLDAvis.gensim
```

```python
# optimal number of topics for reviews on pros = 17, second-optimal number is 9
# will try 17 first and then 9
file = open('/Users/chengchen/glassdoor/model/lda_17_pro.pkl','rb')
lda_model = pickle.load(file)
file.close()
lda_model.print_topics()
```

```python
file = open('/Users/chengchen/glassdoor/data/processed/review_pros_dic_mat.pkl', 'rb')
[dictionary, doc_term_matrix] = pickle.load(file)
file.close()
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, doc_term_matrix, dictionary)
vis
```

```python
# number_of_topics = 9
file = open('/Users/chengchen/glassdoor/model/lda_9_pro.pkl','rb')
lda_model = pickle.load(file)
file.close()
lda_model.print_topics()
```
